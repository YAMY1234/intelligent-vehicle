```python
road_info = dict()
for station in app_platform_info.keys():
    next_station = app_platform_info[station]['next_p_ids'].split(',')
    if ',' in app_platform_info[station]['next_p_distances']:
        next_station_dist = app_platform_info[station]['next_p_distances'].split(',')
    else:
        next_station_dist = app_platform_info[station]['next_p_distances'].split('，')
    road_info[station] = dict()
    for station1 in app_platform_info.keys():
        if station1 == station:
            road_info[station][station1] = dict()
            road_info[station][station1]['dist'] = 0
            road_info[station][station1]['rounte'] = []
        elif station1 not in next_station:
            road_info[station][station1] = dict()
            road_info[station][station1]['dist'] = np.inf
            road_info[station][station1]['route'] = []
        else:
            road_info[station][station1] = dict()
            road_info[station][station1]['dist'] = float(next_station_dist[next_station.index(station1)])
            road_info[station][station1]['route'] = [station]
```

```python
def startwith(start, road_info):
    passed = [start]
    nopass = list(road_info.keys())
    nopass.remove(start)
    dis = road_info[start]
    
    while len(nopass):
        idx = nopass[0]
        for i in nopass:
            if dis[i]['dist'] < dis[idx]['dist']: idx = i
        nopass.remove(idx)
        passed.append(idx)
        
        for i in nopass:
            if dis[idx]['dist'] + road_info[idx][i]['dist'] < dis[i]['dist']:
                dis[i]['dist'] = dis[idx]['dist'] + road_info[idx][i]['dist']
                dis[i]['route'] = dis[idx]['route'] + [idx]
                
    road_info[start] = dis
    return road_info
```

```python
def test_ticket(num, stations):
    tickets = dict()
    usid = 0
    #station1 = stations[random.randint(0, len(stations)-1)]
    #station2 = stations[random.randint(0, len(stations)-1)]
    for i in range(num):
        station1 = stations[random.randint(0, len(stations)-1)]
        station2 = stations[random.randint(0, len(stations)-1)]
        ticket = dict()
        ticket['oId'] = str(i)
        ticket['ticketNumber'] = str(random.randint(1, 36))
        ticket['u_ids'] = [usid+i for i in range(int(ticket['ticketNumber']))]
        ticket['fromId'] = station1
        ticket['toId'] = station2
        ticket['startTime'] = '1234567'
        ticket['orderTime'] = '12345678'
        tickets[str(i)] = ticket
        if random.randint(0, 1) == 1:
            ticket['seat_preference'] = dict()
            length =[1,2]
            width = [3,4]
            for uid in ticket['u_ids']:
                ticket['seat_preference'][str(uid)] = [length[random.randint(0, 1)], width[random.randint(0, 1)]]
        usid += num
    return tickets
```

```python
def merge_to_newtickets(tickets):
    new_tickets = dict()
    odpair = []
    for i in tickets.keys():
        if (tickets[i]['fromId'], tickets[i]['toId']) not in odpair:
            odpair.append((tickets[i]['fromId'], tickets[i]['toId']))
    ticket_group = pd.DataFrame(tickets).transpose().groupby(['fromId', 'toId'])
    for pair in odpair:
        group_temp = ticket_group.get_group(pair)
        new_ticket = dict()
        new_ticket['fromId'] = pair[0]
        new_ticket['toId'] = pair[1]
        new_ticket['ticketNumber'] = group_temp['ticketNumber'].sum()
        new_tickets[pair] = new_ticket
    return new_tickets
new_tickets = merge_to_newtickets(tickets)
```
